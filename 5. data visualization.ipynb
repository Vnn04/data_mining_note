### Data mining Chart

#### Explaratory Data Analysis - EDA
Biểu đồ đơn giản là công cụ cơ bản giúp bạn hiểu dữ liệu một cách trực quan, dễ dàng phát hiện các mẫu và xu hướng. Chúng thường được sử dụng ở bước khám phá dữ liệu ban đầu.

#### 1. Biểu đồ cột (Bar Chart)
**Mục đích**
+ So sánh số lượng hoặc giá trị giữa các danh mục
+ Ví dụ: so sánh doanh thu giữa các cửa hàng hoặc số người đăng ký giữa các khóa học

**Khi nào sử dụng?**
+ Khi cần minh họa giá trị của từng danh mục riêng biệt
+ Phù hợp với dữ liệu phân loại

```python
import matplotlib.pyplot as plt

stores = ['store A', 'store B', 'store C', 'store D']
sales = [100, 150, 80, 200]
```

```python
plt.bar(stores, sales, color='lightblue')
plt.xlabel('Stores')
plt.ylabel('Sales')
plt.title('Sales Comparison by Store')
plt.show()
```

#### 2. Biểu đồ tròn (Pie Chart)
**Mục đích**
+ Thể hiện tỷ lệ phần trăm của các danh mục trong tổng thể.
+ Ví dụ: phân bố thị phần của các công ty trong một ngành
**Khi nào sử dụng?**
+ Khi muốn nhấn mạnh tỷ lệ tương đối giữa các phần tử

```python
labels = ['Company A', 'Company B', 'Company C', 'Company D']
sizes = [30, 25, 25, 20]
```

```python
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Market Share Distribution')
plt.show()
```

**Lưu ý**: Biểu đồ tròn dễ hiểu khi số lượng danh mục ít, nhưng không phù hợp khi có quá nhiều phần tử

#### 3. Biểu đồ đường (line chart)
**Mục đích**
+ Thể hiện xu hướng hoặc sự thay đổi của dữ liệu theo thời gian
+ Ví dụ: theo dõi sự biến động nhiệt độ hàng ngày trong một tháng

**Khi nào sử dụng?**
+ Khi cần nhấn mạnh sự thay đổi liên tục hoặc xu hướng dài hạn

```python
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
temperatures = [30, 32, 33, 31, 29, 28, 30]
```

```python
plt.plot(days, temperatures, marker='o', linestyle='-', color='lightblue')
plt.xlabel('Days')
plt.ylabel('Temperature (°C)')
plt.title('Daily Temperature Trend')
plt.show()
```

**Lưu ý** Biểu đồ đường giúp phát hiện xu hướng và những thay đổi nhỏ qua thời gian, chẳng hạn như nhiệt độ giảm dần từ giữa tuần.

#### 4. Histogram (Biểu đồ tần suất)
**Mục đích**
+ Thể hiện phân phối tần suất của dữ liệu liên tục
+ Ví dụ: xem phân phối điểm thi của học sinh trong một kỳ thi

**Khi nào sử dụng?**
+ Khi bạn muốn hiểu cách dữ liệu phân bố: Dữ liệu có bị lệch, đối xứng hay có nhiều ngoại lệ không?

```python
import numpy as np 

scores = np.random.normal(loc=70, scale=10, size=100)

plt.hist(scores, bins=10, color='purple', alpha=0.7)
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.title('Distribution of Exam Scores')
plt.show()
```

**Lưu ý:** Histogram cho thấy dữ liệu phân bố theo dạng chuông (normal distribution) với điểm số tập trung nhiều quanh giá trị 70

#### 5. Biểu đồ miền (Area Chart)
**Mục đích**
+ Giống biểu đồ đường nhưng khu vực bên dưới được tô màu, giúp nhấn mạnh độ lớn của giá trị theo thời gian
+ Ví dụ: thể hiện lượng mưa tích lũy theo tháng

**Khi nào sử dụng?**
+ Khi cần so sánh xu hướng và độ lớn của một chuỗi dữ liệu qua thời gian

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
rainfall = [30, 45, 60, 80, 100]
```

```python
plt.fill_between(months, rainfall, color='skyblue', alpha=0.5)
plt.plot(months, rainfall, marker='o', color='blue')
plt.xlabel('Months')
plt.ylabel('Rainfall (mm)')
plt.title('Cumulative Rainfall')
plt.show()
```

#### 6. Biểu đồ thanh xếp chồng (Stacked Bar Chart)
**Mục đích**
+ Thể hiện tổng giá trị và cấu trúc của từng phần trong tổng
+ Ví dụ: Thể hiện tổng doanh số bán hàng và đóng góp của từng sản phẩm

**Khi nào sử dụng?**
+ Khi muốn hiển thị nhiều thành phần trong một nhóm và so sánh chúng với các nhóm khác.

```python
months = ['Jan', 'Feb', 'Mar']
product_a = [5, 3, 6]
product_b = [2, 4, 5]
```

```python
plt.bar(months, product_a, label='Product A', color='lightcoral')
plt.bar(months, product_b, bottom=product_a, label='Product B', color='lightskyblue')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.title('Stacked Bar Chart Example')
plt.legend()
plt.show()
```

**Lưu ý**: biểu đồ này cho thấy tổng doanh số mỗi tháng và đóng góp của từng sản phẩm.

#### Trend Analysis Charts
Biểu đồ xu hướng là công cụ hữu hiệu để theo dõi và phân tích sự thay đổi của dữ liệu theo thời gian. Chúng giúp xác định các mẫu (patterns), biến động và dự báo tương lai.

| **Biểu đồ**              | **Mục đích chính**                               | **Ví dụ điển hình**                      |
|--------------------------|--------------------------------------------------|------------------------------------------|
| Line Chart (Biểu Đồ Đường) | Thể hiện biến động qua thời gian                 | Giá cổ phiếu theo ngày                  |
| Area Chart (Biểu Đồ Khu Vực) | Nhấn mạnh độ lớn theo thời gian                 | Lượng mưa tích lũy                       |
| Step Chart (Biểu Đồ Bậc Thang) | Thể hiện biến động với bước nhảy rời rạc       | Giá điện theo khung giờ                 |
| Stream Graph (Biểu Đồ Dòng Chảy) | So sánh nhiều chuỗi dữ liệu động           | Lưu lượng truy cập web theo giờ         |
| Dual-Axis Chart (Biểu Đồ Trục Kép) | So sánh hai loại dữ liệu có đơn vị khác nhau | Doanh thu và số lượt truy cập          |


#### Biểu đồ bậc thang
**Mục đích**: Thể hiện dữ liệu có các bước thay đổi rời rạc
+ Ứng dụng: giá điện theo từng khung giờ, phí giao hàng theo khoảng cách

```python
hours = [0, 6, 12, 18, 24]
electricity_price = [5, 10, 7, 12, 8]
```

```python
plt.step(hours, electricity_price, where='mid', color='green')
plt.xlabel('Hour')
plt.ylabel('Price ($)')
plt.title('Electricity Price by Time of Day')
plt.show()
```

#### Dual-Axis Chart (Biểu đồ trục kép)
**Mục đích**: So sánh hai chuỗi dữ liệu khác đơn vị trên cùng một biểu đồ với hai trục Y.
+ Ứng dụng: So sánh doanh thu và số lượt truy cập theo thời gian.

```python
fig, ax1 = plt.subplots()

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
revenue = [1000, 1200, 1100, 1300, 1400]
visitors = [200, 300, 250, 400, 450]

ax1.set_xlabel('Months')
ax1.set_ylabel('Revenue ($)', color='blue')
ax1.plot(months, revenue, color='blue', marker='o')
ax1.tick_params(axis='y', labelcolor='blue')

ax2 = ax1.twinx()
ax2.set_ylabel("Visitors", color='red')
ax2.plot(months, visitors, color='red', marker='x')
ax2.tick_params(axis='y', labelcolor='red')

plt.title('Revenue vs Visitors Trend')
plt.show()
```

### Biểu đồ tương quan

Biểu đồ tương quan là công cụ trực quan hóa mối quan hệ giữa hai hoặc nhiều biến. Những biểu đồ này giúp bạn nhận biết mức độ và hướng tương quan giữa các biến. Trong Data mining và phân tích dữ liệu, biểu đồ tương quan rất quan trọng để phát hiện các mẫu ẩn, đưa ra dự báo và ra quyết định.

|**Loại biểu đồ**|**Mô tả**|**Ví dụ**|
|-------------------------|---------------------------------------|-----------------------------------|
|Scatter Plot| So sánh hai biến và kiểm tra mối tương quan tuyến tính|Chiều cao và cân nặng|
|Heatmap | Thể hiện ma trận tương quan giữa nhiều biến | Tương quan giữa các biến tài chính |
| Bubble Chart | Giống Scatter Plot nhưng bổ sung thêm kích thước để biểu thị biến thứ ba | Doanh thu, chi phí và lợi nhuận theo từng cửa hàng |
|Pair Plot | Vẽ nhiều Scatter Plots để kiểm tra từng cặp biến trong dataset | Kiểm tra các biến trong dataset Iris|

#### 1. Scatter Plot (Biểu đồ phân tán)
Biểu đồ phân tán giúp bạn kiểm tra mối quan hệ tuyến tính giữa hai biến liên tục
+ Ứng dụng: Phân tích mối quan hệ giữa chiều cao và cân nặng, doanh thu và số lượng bán hàng.

```python
import matplotlib.pyplot as plt 

height = [150, 160, 165, 170, 175, 180]
weight = [50, 55, 60, 65, 70, 75]

plt.scatter(height, weight, color='skyblue', marker='o')
plt.xlabel('Chiều cao (cm)')
plt.ylabel('Cân nặng (kg)')
plt.title('Mối tương quan giữa chiều cao và cân nặng')
plt.show()
```

**Giải thích**: Nếu các điểm dữ liệu tạo thành một đường thẳng tăng dần, có tương quan dương. Nếu dốc xuống, là tương quan âm. Nếu không có mẫu rõ ràng thì không có tương quan.

#### 2. Heatmap 
Heatmap thể hiện ma trận tương quan giữa nhiều biến, với màu sắc hiển thị mức đột ương quan. 
+ Ứng dụng: Phân tích tương quan giữa các biến kinh tế như GDP, tỷ lệ thất nghiệp, lãi suất.

```python
import seaborn as sns 
import pandas as pd 

data = {'Doanh thu': [10, 200, 300, 400, 500],
        'Chi phí': [50, 100, 150, 20, 250],
        'Lợi nhuận': [50, 100, 10, 200, 20]}
df = pd.DataFrame(data)

corr_matrix = df.corr()

sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Ma trận tương quan giữa doanh thu, chi phí và lợi nhuận')
plt.show()
```

**Giải thích**: Các ô được tô màu xanh dương (tương quan âm) đến đỏ (tương quan dương). Giá trị tương quan càng gần 1 hoặc -1 thì mối quan hệ giữa các biến càng mạnh.

#### 3. Bubble Chart
Biểu đồ này giống Scatter Plot, nhưng mỗi điểm dữ liệu là một bong bong có kích thước khác nhau, biểu thị một biến thứ ba
+ Ứng dụng: Phân tích doanh thu, chi phí và lợi nhuận của từng cửa hàng.

```python
revenue = [100, 200, 300, 400, 500]
cost = [50, 100, 150, 200, 250]
profit = [10, 20, 30, 40, 50]  

plt.scatter(revenue, cost, s=[p * 10 for p in profit], color='green', alpha=0.5)
plt.xlabel("Doanh thu ($)")
plt.ylabel('Chi phí ($)')
plt.title('Phân tích Doanh thu, chi phí và lợi nhuận')
plt.show()
```

#### 4. Pair Plot
Pair Plot vẽ tất cả các cặp Scatter Plot trong một dataset, giúp bạn nhanh chóng kiểm tra tương quan giữa các biến.
+ Ứng dụng: phân tích dataset Iris để kiểm tra mối quan hệ giữa các đặc trưng của hoa.

```python
import seaborn as sns 
import matplotlib.pyplot as plt 

df = sns.load_dataset('iris')

sns.pairplot(df, hue='species')
plt.show()
```

**Giải thích**: Pair Plot cho thấy từng cặp biến và mức độ tương quan giữa chúng. Mỗi màu sắc đại diện cho một loài hoa trong dataset.

#### Khi nào sử dụng biểu đồ tương quan?
+ Xác định mối quan hệ giữa các biến: Ví dụ, chiều cao và cân nặng có tương quan không?
+ Lọc ra các biến quan trọng: Nếu hai biến có tương quan cao, bạn có thể loại bỏ một trong số chúng để giảm bớt dư thừa.
+ Xác định các mẫu ẩn: Giúp tìm ra các mẫu tiềm năng trong dữ liệu
+ Dự báo: nếu có mối tương quan mạnh, bạn có thể dự đoán biến này dựa vào biến khác.

### Distribution Charts
Biểu đồ phân phối được sử dụng để trực quan hóa sự phân bố của dữ liệu và giúp bạn hiểu rõ hơn về các giá trị trong tập dữ liệu trải rộng, tập trung hay có dạng bất đối xứng. Những biểu đồ này rất quan trọng trong việc khám phá dữ liệu ví chúng cho thấy đặc điểm thống kê như trung bình, độ lệch chuẩn, phân vị và các giá trị ngoại lệ.

#### Các loại biểu đồ phân phối phổ biến
|Loại biểu đồ | Mục đích | Ví dụ ứng dụng | 
|------------------------|------------------------------|---------------------------------|
|Histogram| Thể hiện số lần xuất hiện của dữ liệu theo từng khoảng | Phân phối chiều cao của học sinh | 
| KDE Plot | Thể hiện mật độ xác suất của dữ liệu liên tục | Phân tích mật độ dân số|
|Box plot | Xác định phân vị, giá trị trung vị và ngoại lệ | So sánh điểm số giữa các lớp | 
|Violin plot | Kết hợp giữa KDE Plot và Box Plot | Phân tích điểm kiểm tra giữa các nhóm |
|Cumulative Distribution function | Thể hiện tỷ lệ tích lũy của dữ liệu | Xác định xác suất tích lũy của biến|

#### 2. KDE Plot 
KDE Plot làm mịn dữ liệu bằng cách ước lượng hàm mật độ xác suất. Đây là dạng biểu đồ mượt hơn Histogram
+ Ứng dụng: Phân tích mật độ dân số hoặc sự phân bố thu nhập

```python
import seaborn as sns 
height = np.random.normal(170, 10, 100)

sns.kdeplot(height, shade=True, color='green')
plt.xlabel('Chiều cao (cm)')
plt.ylabel('Hàm mật độ của chiều cao học sinh')
plt.show()
```

**Giải thích**: Biểu đồ cho thấy mật độ của dữ liệu và giúp phát hiện khu vực tập trung dữ liệu

#### 3. Box Plot
Box Plot hiển thị trung vị, phân vị và giá trị ngoại lệ trong dữ liệu, giúp so sánh sự phân phối giữa nhiều nhóm.
+ Ứng dụng: so sánh điểm kiểm tra giữa các lớp

```python
import seaborn as sns 

data = [7, 8, 9, 9, 10, 12, 13, 15, 17, 18, 20, 30]

sns.boxplot(data=data)
plt.title('Phân phối điểm kiểm tra')
plt.show()
```

**Giải thích**: biểu đồ hiển thị trung vị (median), khoảng tứ phân vị (IQR) và các điểm dữ liệu ngoại lệ outliers

#### 4. Violin Plot
Violin Plot là sự kết hợp giữa KDE Plot và Box Plot giúp biểu diễn cả mật độ dữ liệu và các đặc trưng thống kê.
+ Ứng dụng: so sánh phân phối điểm kiểm tra giữa các lớp học hoặc nhóm

```python
data = sns.load_dataset('tips')

sns.violinplot(x='day', y='total_bill', data=data)
plt.title('Phân phối hóa đơn thanh toán theo ngày')
plt.show()
```

**Giải thích**: Violin Plot cho thấy sự phân bố hóa đơn thanh toán theo ngày và các đặc trưng như trung bị và ngoại lệ.

#### 5. Cumulative Distribution Function (CDF)
Biểu đồ CDF biểu diễn tỷ lệ tích lũy của dữ liệu, giúp bạn biết tỷ lệ phần trăm dữ liệu nằm dưới một giá trị nhất định.
+ Ứng dụng: phân tích xác suất tích lũy của thu nhập hoặc điểm thi.

```python
scores = np.random.normal(70, 10, 100)

count, bins_count = np.histogram(scores, bins=10, density=True)
cdf = np.cumsum(count) / sum(count)

plt.plot(bins_count[1:], cdf, marker='o')
plt.xlabel('Điểm thi')
plt.ylabel('Tỷ lệ tích lũy')
plt.title('CDF của điểm thi')
plt.show()
```

#### Khi nào sử dụng biểu đồ phân phối?
+ Khám phá dữ liệu: Hiểu rõ hơn về đặc điểm và sự phân bố của dữ liệu
+ Phát hiện bất thường: Phát hiện giá trị ngoại lệ hoặc các mẫu bất thường
+ So sánh các nhóm: Kiểm tra sự khác biệt trong phân phối giữa nhiều nhóm.
+ Đưa ra quyết định: Giúp bạn dự báo và lên kế hoạch dựa trên đặc điểm của dữ liệu.

### Advanced Charts
Biểu đồ nâng cao giúp trực quan hóa các tập dữ liệu phức tạp và thể hiện mối quan hệ đa chiều giữa các biến. Những biểu đồ này thường được sử dụng trong khai phá dữ liệu, phân tích dữ liệu lớn, và trực quan hóa những xu hướng hoặc mối quan hệ khó phát hiện.

| **Loại Biểu Đồ**                  | **Mục Đích**                                | **Ví Dụ Ứng Dụng**                          |
|-----------------------------------|---------------------------------------------|---------------------------------------------|
| Heatmap (Biểu Đồ Nhiệt)           | Hiển thị cường độ và mối quan hệ giữa nhiều biến | Ma trận tương quan của các chỉ số tài chính |
| Treemap (Biểu Đồ Cây)             | Trình bày cấu trúc phân cấp của dữ liệu     | Tỷ lệ đóng góp của danh mục đầu tư          |
| Radar Chart (Biểu Đồ Radar)       | So sánh nhiều biến đa chiều trong một đối tượng | Đánh giá hiệu suất nhân viên               |
| Sankey Diagram (Biểu Đồ Sankey)   | Trình bày luồng và sự phân phối dữ liệu     | Dòng chảy năng lượng giữa các hệ thống     |
| Chord Diagram (Biểu Đồ Hợp)       | Thể hiện mối liên hệ qua lại giữa các nhóm dữ liệu | Phân tích luồng dữ liệu giữa các khu vực |
| Bubble Chart (Biểu Đồ Bong Bóng)  | So sánh dữ liệu theo ba biến số            | Phân tích doanh số theo vùng và sản phẩm   |


#### Biểu đồ cây
Treemap giúp trình bày phân cấp của dữ liệu và phân bổ tỷ lệ giữa các phần tử bằng kích thước ô
**Ứng dụng**
+ Trình bày tỷ lệ đóng góp của từng hạng mục trong danh mục đầu tư
+ Phân tích cơ cấu thị trường hoặc công ty con trong tập đoàn

```python
import plotly.express as px

# Dữ liệu mẫu: Thị phần của các sản phẩm
data = dict(
    labels=["Công nghệ", "Y tế", "Ngân hàng", "Năng lượng"],
    parents=["", "", "", ""],
    values=[500, 300, 200, 400]
)

# Vẽ Treemap
fig = px.treemap(data, path=['labels'], values='values', title='Treemap: Thị phần các ngành')
fig.show()
```

#### Radar Chart
Radar Chart so sánh nhiều biến đa chiều cho một đối tượng, hữu ích trong việc phân tích hiệu suất hoặc năng lực
**Ứng dụng**:
+ Đánh giá hiệu suất nhân viên dựa trên nhiều tiêu chí
+ So sánh các sản phẩm dựa trên tính năng

```python
import plotly.express as px

df = px.data.iris().iloc[0:1].melt(id_vars=["species"], var_name="metric", value_name="value")

fig = px.line_polar(df, r='value', theta='metric', line_close=True, title='Radar Chart: Hiệu suất nhân viên')
fig.show()
```

**Giải thích**: Biểu đồ radar thể hiện hiệu suất của nhân viên qua nhiều chỉ số, với mỗi chỉ số là một trục trong hình tròn

#### Sankey Diagram
Biểu đồ Sankey biểu diễn luồng dữ liệu hoặc năng lượng từ nguồn đến đích, với độ rộng của luồng thể hiện số lượng
**Ứng dụng**: 
+ Theo dõi luồng năng lương qua các hệ thống
+ Phân tích luồn khác hàng từ giai đoạn này sang giai đoạn khác

```python
import plotly.graph_objects as go

fig = go.Figure(go.Sankey(
    node=dict(label=["Nguồn", "Đích 1", "Đích 2"]),
    link=dict(source=[0, 0], target=[1, 2], value=[8, 4])
))

fig.update_layout(title_text="Sankey Diagram: Luồng năng lượng", font_size=10)
fig.show()
```

Giải thích: Biểu đồ Sankey thể hiện dòng chảy dữ liệu giữa các nút, với độ rộng của dòng tương ứng với lượng dữ liệu.

#### 6. Bubble chart
Bubble chart mở rộng Scatter Plot bằng cách thêm kích thước bong bóng để thể hiện một biến thứ ba
**Ứng dụng**:
+ Phân tích doanh số bán hàng theo sản phẩm và vùng
+ So sánh GDP, dân số và chỉ số phát triển con người giữa các quốc gia

```python
import plotly.express as px

df = px.data.gapminder().query("year == 2007")

fig = px.scatter(df, x='gdpPercap', y='lifeExp', size='pop', color='continent',
                 hover_name='country', log_x=True, size_max=60)
fig.show()
```

Giải thích: Biểu đồ Bong Bóng cho phép phân tích ba biến số GDP, tuổi thọ, và dân số trên cùng một hình.
